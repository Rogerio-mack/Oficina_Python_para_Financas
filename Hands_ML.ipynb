```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt             
%matplotlib inline
import seaborn as sns
```

# **CASE: Apartment Price** 



```python
df = pd.read_excel('https://github.com/Rogerio-mack/Oficina_Python_para_Financas/raw/main/data/apartamentos_zona_sul.xlsx')
df.head()
```

```python
df.columns
```

```python
import statsmodels.formula.api as sm

lm = sm.ols(formula='preco ~ areaM2', data=df)
lm = lm.fit()
print(lm.summary())
```

```python
df_new = df.iloc[10:11]
df_new
```

```python
lm.predict(df_new['areaM2'])
```

```python
from sklearn import linear_model

X = df[['areaM2']]
y = df['preco']

reg = linear_model.LinearRegression()

reg.fit(X, y)
```

```python
reg.score, reg.coef_, reg.intercept_
```

```python
reg.predict(df_new[['areaM2']])
```

# **CASE: Car Sales** 

```python
df = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/MASS/Cars93.csv',index_col=0)
df = df.reset_index(drop=True)
df.head()
```

```python
df.columns
```

```python
from sklearn.linear_model import LogisticRegression

# Entradas e Saídas
X = df[['Price', 'MPG.city', 'EngineSize', 'Horsepower', 'RPM', 'Length', 'Wheelbase', 'Width', 'Weight']]
y = df['Origin']

# Definição
clf = LogisticRegression(max_iter=1000)

# Treinamento
clf.fit(X,y)

# Avaliação
y_pred = clf.predict(X)

print(clf.score(X,y))
```

```python
df['y_pred'] = y_pred
df[['Manufacturer', 'Model', 'Type', 'Price', 'Max.Price', 'Horsepower', 'RPM', 'Origin', 'y_pred']]
```

```python
df[['Manufacturer', 'Model', 'Type', 'Price', 'Max.Price', 'Horsepower', 'RPM', 'Origin', 'y_pred']][ df.Origin != df.y_pred ]
```
